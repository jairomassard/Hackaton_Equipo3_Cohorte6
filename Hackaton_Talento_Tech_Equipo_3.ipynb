### **1. Extraer los datos del archivo SQL en Colab y convertirlo en un archivo .cvs**

```python
import re
import pandas as pd

# Definir la ruta del archivo SQL
sql_file_path = '/content/drive/MyDrive/hackathon_equipo3/data_inicial/personas_desaparecidas.sql'

# Leer el archivo SQL
with open(sql_file_path, 'r', encoding='utf-8') as sql_file:
    sql_script = sql_file.read()

# Expresión regular para encontrar los valores en las sentencias INSERT
insert_pattern = r"INSERT INTO `personas_desaparecidas`.*?VALUES\s*(.*);"

# Encontrar todas las sentencias de INSERT
matches = re.findall(insert_pattern, sql_script, re.DOTALL)

# Lista donde almacenaremos las filas de datos
rows = []

# Procesar cada coincidencia para extraer los valores
for match in matches:
    # Limpiar y dividir los valores en cada fila teniendo en cuenta comillas
    values = re.findall(r"\(([^)]+)\)", match)
    for value in values:
        # Separar los valores usando una expresión regular que respete las comas dentro de comillas
        row = re.findall(r"'([^']*)'|(\d+)", value)
        # Limpiar los resultados y organizar los valores
        cleaned_row = [item[0] if item[0] else item[1] for item in row]
        rows.append(cleaned_row)

# Definir las columnas de la tabla
columns = ['id', 'nombre_desaparecido', 'documento_desaparecido', 'fecha_nacimiento', 'edad', 'municipio', 'departamento', 'relato_desaparicion', 'fecha_desaparicion', 'lugar_desaparicion', 'origen_relato']

# Crear un DataFrame con los datos extraídos
df = pd.DataFrame(rows, columns=columns)

# Guardar el DataFrame en un archivo CSV
csv_file_path = '/content/drive/MyDrive/hackathon_equipo3/personas_desaparecidas.csv'
df.to_csv(csv_file_path, index=False, encoding='utf-8')

print(f"Archivo CSV guardado en: {csv_file_path}")

# Mostrar las primeras filas del DataFrame
df.head()

```

### **1. Extracción de Variables sobre el archivo de relatos original**

```python
# Cargar el archivo CSV original (sin anonimizar)
df_relatos_word_original = pd.read_csv('/content/drive/MyDrive/hackathon_equipo3/relatos_no_anonimizados.csv')

# Verificar la carga del DataFrame original
df_relatos_word_original.head()
```

```python
import spacy
import re    # segmenta la data y busca patronmes

# Cargar el modelo de español de spaCy
nlp = spacy.load("es_core_news_md")

# Funciones para extraer variables clave

def extract_dates(text):
    # Buscar fechas en el formato 'YYYY-MM-DD' o 'DD de Mes de YYYY'
    date_pattern = r"\b\d{4}-\d{2}-\d{2}\b|\b\d{1,2}\sde\s\w+\sde\s\d{4}\b"
    return re.findall(date_pattern, text)

def extract_groups(text):
    # Buscar menciones de grupos armados
    group_pattern = r"\b(guerrilla|paramilitar|grupo armado|combatientes|militares|autodefensas)\b"
    return re.findall(group_pattern, text)

def extract_activities(text):
    # Buscar actividades o eventos clave como desaparición, secuestro, etc.
    activity_pattern = r"\b(desapareció|secuestraron|desplazaron|asesinato|bombardeo|detenido|retenido|capturado)\b"
    return re.findall(activity_pattern, text)

# Función para extraer todas las variables de interés
def extraer_variables_categorizar(text):
    # Procesar el texto con spaCy
    doc = nlp(text)

    # Extraer nombres y lugares
    personas = [ent.text for ent in doc.ents if ent.label_ == "PER"]
    lugares = [ent.text for ent in doc.ents if ent.label_ == "LOC"]

    # Aplicar reglas para extraer fechas, grupos armados y actividades
    fechas = extract_dates(text)
    grupos = extract_groups(text)
    actividades = extract_activities(text)

    # Categorizar la información extraída
    return {
        "personas_involucradas": personas,
        "lugares_mencionados": lugares,
        "fechas_importantes": fechas,
        "grupos_armados_mencionados": grupos,
        "eventos_actividades": actividades
    }

# Aplicar la extracción sobre el DataFrame de relatos no anonimizados
df_relatos_word_original['variables_extraidas'] = df_relatos_word_original['relato'].apply(extraer_variables_categorizar)

# Mostrar las variables extraídas de los primeros relatos
df_relatos_word_original[['relato', 'variables_extraidas']].head(10)

```

```python
import pandas as pd

# Cargar el archivo personas_desaparecidas_limpio.csv
file_path = '/content/drive/MyDrive/hackathon_equipo3/personas_desaparecidas_limpio.csv'
df_personas_desaparecidas = pd.read_csv(file_path)

# Mostrar las primeras filas del DataFrame para verificar la carga
df_personas_desaparecidas.head()
```

```python
import spacy
import re

# Cargar el modelo de español de spaCy
nlp = spacy.load("es_core_news_md")

# Funciones para extraer variables clave adicionales

def extract_groups(text):
    # Ampliamos los términos que representan grupos armados
    group_pattern = r"\b(guerrilla|paramilitar|grupo armado|militares|autodefensas|hombres armados|hombres vestidos de camuflado|grupo guerrillero|soldados)\b"
    return re.findall(group_pattern, text)

def extract_activities(text):
    # Ampliamos los términos que representan actividades o eventos clave
    activity_pattern = r"\b(desapareció|secuestraron|desplazaron|asesinato|bombardeo|detenido|retenido|capturado|llevado|sacada|interrogado|encarcelado|golpeado|incursión|operación militar)\b"
    return re.findall(activity_pattern, text)

# Función para extraer todas las variables clave de la columna relato_desaparicion
def extraer_variables_desaparicion(text):
    doc = nlp(text)

    # Extraer entidades reconocidas por spaCy
    personas = [ent.text for ent in doc.ents if ent.label_ == "PER"]
    lugares = [ent.text for ent in doc.ents if ent.label_ == "LOC"]
    fechas = [ent.text for ent in doc.ents if ent.label_ == "DATE"]

    # Aplicar reglas para extraer grupos armados y actividades
    grupos = extract_groups(text)
    actividades = extract_activities(text)

    return {
        "personas_involucradas": personas,
        "lugares_mencionados": lugares,
        "fechas_importantes": fechas,
        "grupos_armados_mencionados": grupos,
        "eventos_actividades": actividades
    }

# Aplicar la extracción sobre la columna "relato_desaparicion"
df_personas_desaparecidas['variables_extraidas'] = df_personas_desaparecidas['relato_desaparicion'].apply(extraer_variables_desaparicion)

# Guardar el archivo con las variables extraídas
df_personas_desaparecidas.to_csv('/content/drive/MyDrive/hackathon_equipo3/personas_desaparecidas_con_variables.csv', index=False)

# Mostrar algunos resultados
df_personas_desaparecidas[['relato_desaparicion', 'variables_extraidas']].head()

```
